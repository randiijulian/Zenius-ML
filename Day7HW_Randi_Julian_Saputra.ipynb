##### Nomor 3a:

Judul dari plot, serta judul dari setiap sumbu, bebas. Tidak harus persis sama, asalkan tidak dibiarkan kosong atau dibikin default (masih pakai underscore, huruf kecil semua, dst). 

```python
fig, (ax1, ax2) = plt.subplots(2, figsize = (9,12))

sns.scatterplot(x="body_mass_g", y="flipper_length_mm", hue="island", palette="muted", ax=ax1,data=df3)

ax1.set_xlabel('')
ax1.set_ylabel('Flipper Length (in mm)')
ax1.set_title('Body Mass of Penguin vs Flipper Length')

sns.scatterplot(x="body_mass_g", y="bill_length_mm", hue="island", palette="muted", ax=ax2,data=df3)

ax2.set_xlabel('Mass of Penguin (in grams)')
ax2.set_ylabel('Bill Length (in mm)')
ax2.set_title('Body Mass of Penguin vs Bill Length')

plt.show()
```

##### Nomor 3b:
Kolom `body_mass_g` lebih baik digunakan untuk memprediksi `flipper_length_mm`.
