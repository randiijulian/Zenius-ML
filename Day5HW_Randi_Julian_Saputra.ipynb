```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import norm
```

##1. Creating Dataframe from Dictionary of List

```python
df = pd.DataFrame({
    "name": ["Anna", "Dane", "David", "Kevin", "Joe", "Rose"], #Columns 1
    "weight": [45.2, 50.5, 62.1, 64.0, 59.0, np.nan], #Columns 2
    "height": [155, 160, 162, 170, 167, 159], #Columns 3
    "birth_year": [2000, 1997, 1996, 1987, 1998, 2003], #Columns 4
    "gender": ["female", "male", "male", "male", "male", "female"], #Columns 5
    "city": ["New York", "Chicago", "Austin", "New York", np.nan, "New York"] #Columns 6
})
df
```

##5. Print all rows that match either of the following conditions

    b. Women with height > 157

```python
df[((df["gender"] == "male") & (df["birth_year"]>1996)) | ((df["gender"] == "female") & (df["height"]>157))]
```

##6. Fill NaN of weight with mean of the gender and fill NaN of city with mode

```python
#mengisi nilai kosong dg pertama2 menggunakan groupby "gender" untuk menghitung rata2
df["weight"]=df["weight"].fillna(df.groupby(['gender'])['weight'].transform('mean'))
#mengisi nilai kosong kolom "city" dengan mode/modus di kolom "city"
df["city"] = df["city"].fillna(df["city"].mode()[0])
df
```

##7. Check outliers of weight and remove if any

```python
#menghitung nilai manual iqr untuk identifikasi outlier

q1 = df["weight"].quantile(0.25) #rumus menghitung q1
q3 = df["weight"].quantile(0.75) #rumus menghitung q3

iqr = q3-q1 #rumus menghitung Interquartile range (iqr)
fence_low  = q1-1.5*iqr #menghitung batas bawah 
fence_high = q3+1.5*iqr #menghitung batas atas


#menampilkan baris "df" dimana pada kolom "weight" terdapat nilai yang kurang dari batas bawah ATAU lebih dari batas atas
df.loc[(df["weight"] < fence_low) | (df["weight"] > fence_high)] 
```
